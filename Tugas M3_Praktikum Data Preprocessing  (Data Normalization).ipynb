```python
import pandas as pd

dataset = pd.read_csv("D:/PENS/Semester 4/Data Mining/titanic.csv")
dataset
```

```python
rows, cols = dataset.shape

print("Jumlah baris: ", rows)
print("Jumlah kolom: ", cols)
```

```python
data = dataset.loc[:,["Age", "Fare"]]
data
```

```python
Class = dataset.loc[:, ["Survived"]]
Class
```

```python
gabung_data = pd.concat([data, Class], axis=1)
gabung_data
```

```python
gabung_data.isnull().sum()
```

```python
mean_age_by_survived = gabung_data.groupby('Survived')['Age'].mean()
mean_age_by_survived
```

```python
for survived, mean_age in mean_age_by_survived.items():
    gabung_data.loc[(gabung_data['Survived'] == survived) & (gabung_data['Age'].isnull()), 'Age'] = mean_age

gabung_data
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

# Normalisasi data
normalized_data = pd.DataFrame(scaler.fit_transform(data), columns=['Age', 'Fare'])
normalized_data
```

```python
#Normalization_ Min-Max 0-1

# With Algorithm
mindata = data.min()
maxdata = data.max()
newmin = 0
newmax = 1

normalized_data_algorithm = ((data - mindata) * (newmax - newmin) / (maxdata - mindata)) + newmin
normalized_data_algorithm.rename(columns={'Age': 'algorithm_Age', 'Fare': 'algorithm_Fare'}, inplace=True)

# With Library
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

normalized_data_library = pd.DataFrame(scaler.fit_transform(data), columns=['Library_Age', 'Library_Fare'])


# merge 2 tables
normalized_data_combined = normalized_data_algorithm.join(normalized_data_library)

# Display the merged table
normalized_data_combined
```

```python
from scipy import stats
from scipy.stats import zscore

# Menghitung z-score dan melakukan normalisasi
normalized_data = zscore(data, nan_policy='omit')
normalized_data
```

```python
#Normalization_ Z-Score

# With Algorithm
import numpy as np

mean_data = np.mean(data, axis=0)
std_data = np.std(data, axis=0)

zscore_data_algorithm = (data - mean_data) / std_data
zscore_data_algorithm.rename(columns={'Age': 'Algorithm_Age', 'Fare': 'Algorithm_Fare'}, inplace=True)

# With Library
from scipy.stats import zscore

zscore_data_library = zscore(data)
zscore_data_library.rename(columns={'Age': 'Library_Age', 'Fare': 'Library_Fare'}, inplace=True)

# Combine both results
normalized_data_combined = zscore_data_algorithm.join(zscore_data_library)
normalized_data_combined
```
