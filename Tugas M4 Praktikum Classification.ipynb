```python

titanic_testlabel = pd.read_csv("D:/PENS/Semester 4/Data Mining/titanic_testlabel.csv")

import numpy as np
import pandas as pd

test_label = []

for index in pos_missing_test1[0]:
    if index in titanic_testlabel.index: 
        titanic_testlabel.drop(index, inplace=True)
        
for index in pos_missing_test2[0]:
    if index in titanic_testlabel.index:  
        titanic_testlabel.drop(index, inplace=True)

for index, row in titanic_testlabel.iterrows():
    test_label.append(row['Survived'])

test_label = pd.DataFrame(test_label, columns=['Survived'] )
print(test_label)
```

```python
#normalisasi pada train_data

#Normalization_ Min-Max 0-1
mindata = train_data.min()
maxdata = train_data.max()
newmin = 0
newmax = 1

train_data= ((train_data-mindata)*(newmax-newmin)/(maxdata-mindata)) + newmin
train_data
```

```python
#normalisai pada test data

#Normalization_ Min-Max 0-1
mindata = train_data.min()
maxdata = train_data.max()
newmin = 0
newmax = 1

test_data= ((test_data-mindata)*(newmax-newmin)/(maxdata-mindata)) + newmin
test_data
```

```python
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Memisahkan fitur dan label dari data latih dan data uji
X_train = train_data
y_train = train_label
X_test = test_data
y_test = test_label

# Inisialisasi dictionary untuk menyimpan hasil akurasi
class_result = {}

# Melakukan klasifikasi dengan k-NN untuk setiap nilai k dari 1 hingga 10
for k in range(1, 11):
    # Membuat model k-NN
    knn = KNeighborsClassifier(n_neighbors=k)
    
    # Melatih model menggunakan data latih
    knn.fit(X_train, y_train.values.ravel())
    
    # Melakukan prediksi pada data uji
    y_pred = knn.predict(X_test)
    
    # Menghitung akurasi dan menyimpannya dalam dictionary
    accuracy = accuracy_score(y_test, y_pred)
    class_result[f'k={k}'] = accuracy

# Menampilkan hasil akurasi untuk setiap nilai k
for k, accuracy in class_result.items():
    print(f'Accuracy for {k}: {accuracy}')
```

```python
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

X_train = train_data
y_train = train_label
X_test = test_data
y_test = test_label

class_result = {}

for k in range(1, 11):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train.values.ravel())
    y_pred = knn.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    
    class_result[f'k={k}'] = accuracy

for k, accuracy in class_result.items():
    print(f'Accuracy for {k}: {accuracy}')
```
