```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split, cross_val_score, KFold, LeaveOneOut
from sklearn.linear_model import LogisticRegression  
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score 
```

```python
import pandas as pd
dataset = pd.read_csv("D:/PENS/Semester 4/Data Mining/titanic.csv")
dataset
```

```python
# Ambil dataset kolom fitur (Sex, Age, Pclass, Fare)
train_data = dataset[['Sex', 'Age', 'Pclass', 'Fare']]

# Lakukan pengisian missing value pada fitur Age dengan nilai mean dari masing-masing class
avg_age_class = train_data.groupby('Pclass')['Age'].mean()

for pclass, mean_age in avg_age_class.items():
    train_data.loc[(train_data['Pclass'] == pclass) & (train_data['Age'].isnull()), 'Age'] = mean_age

train_data['Fare'].fillna(train_data['Fare'].mean(), inplace=True)
train_data
```

```python
# Ubah fitur kategorikal 'Sex' menjadi numerikal
train_data.loc[:, 'Sex'] = train_data['Sex'].replace({'male': 0, 'female': 1})
train_data
```

```python
label = dataset.loc[:, ['Survived']]
label
```

```python
#normalisasi pada train_data

#Normalization_ Min-Max 0-1
# Inisialisasi MinMaxScaler
scaler = MinMaxScaler()

# Memilih fitur 'Age' dan 'Fare' untuk dinormalisasi
train_data[['Age', 'Fare']] = scaler.fit_transform(train_data[['Age', 'Fare']])

# Menampilkan train_data setelah transformasi
train_data
```

```python
# Inisialisasi KNeighborsClassifier dengan k=3
knn = KNeighborsClassifier(n_neighbors=3)

# Split data menjadi train dan test set (Hold-out Method)
X_train, X_test, y_train, y_test = train_test_split(train_data, label, test_size=0.3, random_state=42)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)

# Menghitung akurasi pada test set
hold_out_error_ratio = 1 - accuracy_score(y_test, y_pred)
print("Hold-out Method Error Ratio:", hold_out_error_ratio)

# K-Fold Cross Validation (k=10)
kf = KFold(n_splits=10, shuffle=True, random_state=42)
kfold_scores = cross_val_score(knn, train_data, label, cv=kf)
kfold_error_ratio = 1 - kfold_scores.mean()
print("K-Fold Error Ratio:", kfold_error_ratio)

# Leave-One-Out (LOO)
loo = LeaveOneOut()
loo_scores = cross_val_score(knn, train_data, label, cv=loo)
loo_error_ratio = 1 - loo_scores.mean()
print("Leave-One-Out (LOO) Error Ratio:", loo_error_ratio)
```
